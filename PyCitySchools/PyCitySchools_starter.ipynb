```python
# Dependencies and Setup
import pandas as pd
import os

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Group Columns by schools
School_Names = school_data_complete.set_index("school_name").groupby(["school_name"])

# Identify School type by School
School_Types = school_data.set_index("school_name")["type"]

# Identify Total Students per School
Students_Per_School = School_Names['Student ID'].count()

# Identify the Total School Budget per School
School_Budget = (school_data.set_index("school_name")["budget"])

# Calculate the Student Budget per Student by School
Budget_Per_Student = (school_data.set_index("school_name")["budget"]/school_data.set_index("school_name")["size"])

# Calculate the Average Math Score by School
Average_Math_Score = School_Names["math_score"].mean()

# Calculate the Average Reading Score by School
Average_Read_Score = School_Names["reading_score"].mean()

# Calculate the % Passing Math per School
Passing_Math_Per = (school_data_complete[school_data_complete["math_score"] >= 70].groupby("school_name")
                       ["Student ID"].count()/Students_Per_School)*100 

# Calculate the % Passing Reading per School
Passing_Read_Per = (school_data_complete[school_data_complete["reading_score"] >= 70].groupby("school_name")
                        ["Student ID"].count()/Students_Per_School)*100

# Calculate the Overall Percent Passing per School
Overall_Pass_Per = (school_data_complete[(school_data_complete["reading_score"] >= 70) & 
                                         (school_data_complete["math_score"] >= 70)].groupby("school_name")
                    ["Student ID"].count()/Students_Per_School)*100                                              

# Create a new dataframe to display the School Summary
School_dfSummary = pd.DataFrame({"School Type": School_Types,
                                 "Total Students": Students_Per_School,
                                 "Total School Budget": School_Budget,
                                 "Per Student Budget": Budget_Per_Student,
                                 "Average Math Score": Average_Math_Score,
                                 "Average Reading Score": Average_Read_Score,
                                 "% Passing Math": Passing_Math_Per,
                                 "% Passing Reading": Passing_Read_Per, 
                                 "% Passing Overall": Overall_Pass_Per})

# Format Columns
School_dfSummary["Total Students"] = School_dfSummary["Total Students"].map("{:,.0f}".format)
School_dfSummary["Total School Budget"] = School_dfSummary["Total School Budget"].map("${:,.0f}".format)
School_dfSummary["Per Student Budget"] = School_dfSummary["Per Student Budget"].map("${:,.2f}".format)

School_dfSummary
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Create Bins and Name them
spending_bins = [0, 584, 629, 644, 675]
group_names = ["<$584", "$585-629", "$630-644", "$645-675"]

school_data_complete["spending_bins"] = (pd.cut(school_data_complete["budget"]/school_data_complete["size"], 
                                                spending_bins, labels = group_names))

# Groupby spening bins
groupby_spending = school_data_complete.groupby("spending_bins")

# Calculate average math score
math_average = groupby_spending["math_score"].mean()

# Calculate average reading score
reading_average = groupby_spending["reading_score"].mean()

# Calculate % passing math
percent_pass_math = (school_data_complete[school_data_complete["math_score"] >=70].groupby("spending_bins")
                     ["Student ID"].count()/groupby_spending["Student ID"].count())*100
# Calculate % passing reading
percent_pass_read = (school_data_complete[school_data_complete["reading_score"] >=70].groupby("spending_bins")
                     ["Student ID"].count()/groupby_spending["Student ID"].count())*100

# Calculate overall passing rate
overall_pass = (school_data_complete[(school_data_complete["reading_score"] >= 70) & 
                                         (school_data_complete["math_score"] >= 70)].groupby("spending_bins")
                    ["Student ID"].count()/groupby_spending["Student ID"].count())*100
                        
scores_by_school_spend = pd.DataFrame({"Average Math Score": math_average, "Average Reading Score": reading_average,
                                       "% Passing Math": percent_pass_math, "% Passing Reading": percent_pass_read,
                                       "% Overall Passing": overall_pass})

# Formatting
scores_by_school_spend.index.name = "Spending Ranges (Per Student)"
scores_by_school_spend["Average Math Score"] = scores_by_school_spend["Average Math Score"].map("{:,.2f}".format)
scores_by_school_spend["Average Reading Score"] = (scores_by_school_spend["Average Reading Score"].map("{:,.2f}".
                                                                                                       format))
scores_by_school_spend["% Passing Math"] = scores_by_school_spend["% Passing Math"].map("{:,.2f}".format)
scores_by_school_spend["% Passing Reading"] = scores_by_school_spend["% Passing Reading"].map("{:,.2f}".format)
scores_by_school_spend["% Overall Passing"] = (scores_by_school_spend["% Overall Passing"].map("{:,.2f}".
                                                                                                         format))

scores_by_school_spend
```
